# vac2rho

This is an iPython notebook to calculate the frequency spectrum from the velocity autocorrelation function.

First: we  start loading the modules needed to run:

```python
%matplotlib inline
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
from Mara.IO import XVGIO
from scipy.special import k1
from scipy.integrate import trapz
from numpy import linalg
```

Now, we load the VACF from a xmgrace file, using the xvg I/O library:

```python
#xvgfile = XVGIO('/home/nacho/doctorado_neutrones/2014/evaluation/d2o/base/vac-h.xvg')
xvgfile = XVGIO('vac-base-h.xvg')
key, data = xvgfile.read()
key = [float(i) for i in key]
```

And plot it:

```python
plt.plot(key, data[0], label='VACF')
plt.xlim(0,3.0)
plt.legend()
plt.xlabel('t [ps]')
plt.ylabel('VACF')
```

Now, we select only the data in the range [0.0, 1.0] picoseconds:

```python
t = array(key)
vac = array(data[0])
t1 = t[np.where(t<1.0)]
vac1 = vac[np.where(t<1.0)]
```

And compute the frequency spectrum by the Fourier cosine transform:

```python
T=298
m=20.87469
hbar=0.6582119
k=0.086173324
hw = linspace(0.0,400.0,2000)
rho = []
for o in hw/hbar:
    r = trapz(cos(o*t1)*vac1, t1)/hbar
    rho.append(r)
rho = array(rho)
rho=rho/trapz(rho, hw)
```

and plot the spectrum:

```python
plot(hw, rho, label='rho')
plt.legend()
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
```

As we can see, the Fourier transform is smooth. If we select a wider time range, we start picking up noise: 

```python
t2 = t[np.where(t<10.0)]
vac2 = vac[np.where(t<10.0)]
hw2 = linspace(0.0,400.0,2000)
rho2 = []
for o in hw2/hbar:
    r = trapz(cos(o*t2)*vac2, t2)/hbar
    rho2.append(r)
rho2 = array(rho2)
rho2=rho2/trapz(rho2, hw2)
plot(hw, rho, label='rho')
plot(hw2, rho2, label='rho2')
plt.legend()
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
xlim(0,10)
ylim(0,0.006)
```

Now, we have to substract the diffusion component. First, we calculate the diffusion coefficient:

```python
D =  rho[0]/(2*m/(pi*k*T*hbar))
wt = 20.87469/(6.5*(20.87469*2+165.7756)) # From experiments by Novikov: translational weight = 6.5 molecules
print "D = ", D, "nm^2*sec^-1"
```

Then we calculate the diffusion spectrum using the Egelstaff-Schofield model and plot it:

```python
def egelstaff_schofield(beta, wt, c):
    if beta < 1e-20:
        es = 2.0*c*wt/pi
    else:
        es = 4.0*c*wt/pi*sqrt(c**2+0.25)*sinh(beta/2.0)*k1(sqrt(c**2+0.25)*beta)
    return es
```

```python
c = (m*D)/(wt*hbar)
print c
beta = hw/(k*T)
rho_diff = [egelstaff_schofield(be,wt,c)/(k*T) for be in beta] # If clause in E-S function does not allow evaluation on lists
```

```python
plot(hw,rho_diff, label='rho_diff')
print rho_diff[0]
xlim(0,50)
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
plt.legend()
```

The solid-like component of the spectrum is the difference:

```python
rho_no_diff = rho- rho_diff
```

```python
plt.plot(hw,rho_no_diff, label='rho_no_diff')
plt.xlim(0,400)
plt.ylim(0,0.012)
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
plt.legend()
```

which now is parabolic:

```python
plt.plot(hw,rho, label='rho')
plt.plot(hw,rho_diff, label='rho_diff')
plt.plot(hw,rho_no_diff, label='rho_no_diff')
plt.xlim(0,10)
plt.ylim(0,0.01)
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
plt.legend()
```

This can be verified by plotting $\sqrt{\rho-\rho_\text{diff}}$ and comparing it with a linear fit:

```python
rho_no_diff[0] = 0
y = sqrt(rho_no_diff[where(hw < 1.5)]) # Linear fit of the square root
x = hw[where(hw < 1.5)]
A = array([ x, ones(len(x))])
param = linalg.lstsq(A.T,y)[0]
plt.plot(hw,sqrt(rho_no_diff), label='sqrt(rho_no_diff)')
plt.plot(hw, param[1] + param[0]*hw, label='linear fit')
plt.xlim(0,10)
plt.ylim(0,sqrt(0.01))
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
plt.legend()
```

```python
def debye_waller_integral(e,rho,kT):
    integrand = [0.0]*len(e)
    integrand[0] = 2*kT**2*rho[1]/e[1]**2
    integrand[1:-1] = kT*rho[1:-1]/(e[1:-1]*tanh(e[1:-1]/(2.0*kT)))
    return trapz(integrand, e)
```

```python
dw = debye_waller_integral(hw,rho_no_diff, k*T)
wc = trapz(rho_no_diff, hw)
print "Debye Waller integral =", dw
print "Normalization of the continuous spectrum =", wc
print "Normalization of translational spectrum =", wt
print "Sum = ", wc + wt
```
