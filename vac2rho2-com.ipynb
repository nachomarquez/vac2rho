# vac2rho

This is an iPython notebook to calculate the frequency spectrum from the velocity autocorrelation function.

First: we  start loading the modules needed to run:

```python
%matplotlib inline
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
from Mara.IO import XVGIO
from scipy.special import k1
from scipy.integrate import trapz
from numpy import linalg
```

Now, we load the VACF from a xmgrace file, using the xvg I/O library:

```python
#xvgfile = XVGIO('/home/nacho/doctorado_neutrones/2014/evaluation/d2o/base/vac-h.xvg')
xvgfile = XVGIO('vac-573-corto-com.xvg')
key, data = xvgfile.read()
key = [float(i) for i in key]
```

And plot it:

```python
plt.semilogy(key, data[0], label='VACF')
plt.xlim(0.0,5.1)
plt.legend()
plt.xlabel('t [ps]')
plt.ylabel('VACF')
```

Now, we select only the data in the range [0.0, 1.0] picoseconds:

```python
t = array(key)
vac = array(data[0])
t1 = t[np.where(t<1.0)]
vac1 = vac[np.where(t<1.0)]
```

And compute the frequency spectrum by the Fourier cosine transform:

```python
hbar=0.6582119
hw = linspace(0.0,50.0,500)
rho = []
for o in hw/hbar:
    r = trapz(cos(o*t1)*vac1, t1)/hbar
    rho.append(r)
rho = array(rho)
rho=rho/trapz(rho, hw)
```

and plot the spectrum:

```python
plot(hw, rho, label='rho')
plt.legend()
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
xlim(0.0, 10.0)
```

As we can see, the Fourier transform is smooth. If we select a wider time range, we start picking up noise: 

```python
t2 = t[np.where(t<10.0)]
vac2 = vac[np.where(t<10.0)]
hw2 = linspace(0.0,50.0,1000)
rho2 = []
for o in hw2/hbar:
    r = trapz(cos(o*t2)*vac2, t2)/hbar
    rho2.append(r)
rho2 = array(rho2)
rho2=rho2/trapz(rho2, hw2)
plot(hw, rho, label='rho')
plot(hw2, rho2, label='rho2')
plt.legend()
plt.xlabel('E [meV]')
plt.ylabel('rho [meV^-1]')
xlim(0,10)
# ylim(0,0.006)
```
