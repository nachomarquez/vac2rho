```python
def egelstaff_schofield(beta, wt, c):
    if beta < 1e-10:
        es = 2.0*c*wt/pi
    else:
        es = 4.0*c*wt/pi*sqrt(c**2+0.25)*sinh(beta/2.0)*k1(sqrt(c**2+0.25)*beta)
    return es
```
