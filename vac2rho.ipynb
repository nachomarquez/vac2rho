```python
%pylab
from Mara.IO import XVGIO
from scipy.fftpack import fft
from supsmu import supsmu
from scipy.special import k1
from scipy.integrate import trapz, quad
```

```python
xvgfile = XVGIO('/home/nacho/doctorado_neutrones/2014/evaluation/d2o/base/vac-h.xvg')
key, data = xvgfile.read()
key = [float(i) for i in key]
```

```python
plt.plot(key, data[0])
plt.xlim(0,1.0)
```

```python
t = np.concatenate([-np.flipud(key[1:-1]), key])
```

```python
vac = np.concatenate([np.flipud(data[0][1:-1]), data[0]]) # Symmetrize the VACF to calculate the cosine Fourier transform with FFT
```

```python
plt.plot(t, vac)
plt.xlim(-1.0, 1.0)
```

```python
dt = t[1]-t[0]
```

```python
rho = np.abs(np.real(fft(vac)))*dt
rho = rho[0:len(rho)]
T=298
m=20.87469
hbar=0.6582119
k=0.086173324
rho=rho*m/(3*pi*k*T*hbar)
```

```python
nu = np.linspace(0,1.0/dt,len(t))
```

```python
e = nu/.2418 # Conversion from THz to meV
```

```python
rho = rho[np.where(e<500)] # We keep the first part of the spectrum
e = e[np.where(e<500)]
```

```python
plt.plot(e,rho)
plt.xlim(0,400)
```

```python
ones = np.ones(len(e)).astype(np.float32)
rho_smooth2 = np.zeros(len(e)).astype(np.float32)
scratch = np.zeros([len(e),7]).astype(np.float32)
supsmu(e,rho,ones, iper=1, span=0.01, alpha=0.0, smo=rho_smooth2, sc=scratch) # Call to Supersmoother (FORTRAN subroutine)
w = trapz(rho_smooth2, e)
rho_smooth2 = rho_smooth2/w
```

```python
plt.plot(e,rho_smooth2)
plt.xlim(0,400)
```

```python
D =  rho_smooth2[0]/(2*m/(pi*k*T*hbar))
wt = 20.87469/(6.5*(20.87469*2+165.7756)) # From experiments by Novikov: translational weight = 6.5 molecules
```

```python
def egelstaff_schofield(beta, wt, c):
    if beta < 1e-10:
        es = 2.0*c*wt/pi
    else:
        es = 4.0*c*wt/pi*sqrt(c**2+0.25)*sinh(beta/2.0)*k1(sqrt(c**2+0.25)*beta)
    return es
```

```python
c = (m*D)/(wt*hbar)
beta = e/(k*T)
rho_diff = [egelstaff_schofield(be,wt,c)/(k*T) for be in beta] # If clause in E-S function does not allow evaluation on lists
```

```python
plot(e,rho_diff)
print rho_diff[0]
xlim(0,50)
```

```python
rho_no_diff = rho_smooth2- rho_diff
```

```python
plt.plot(e,rho_no_diff)
plt.xlim(0,400)
plt.ylim(0,0.012)
```

```python
plt.plot(e,rho)
plt.plot(e,rho_smooth2)
plt.plot(e,rho_no_diff)
plt.xlim(0,10)
plt.ylim(0,0.01)
```

```python
e[0] = 1e-20
e_cut = 2.0
w1 = trapz(rho_no_diff[np.where(e<e_cut)], e[np.where(e<e_cut)])
rho_no_diff2 = rho_no_diff.copy()
maxi = np.where(e<e_cut)[-1][-1]
for i in np.where(e<e_cut)[-1]:
    rho_no_diff2[i] = 3.0*e[i]**2/e[maxi]**3*w1
rho_no_diff2 = rho_no_diff2/trapz(rho_no_diff2, e)*(1-wt)
plt.plot(e,rho_no_diff)
plt.plot(e,rho_no_diff2)
plt.xlim(0,10)
plt.ylim(0,0.01)
```

```python
def debye_waller_integral(e,rho,kT):
    integrand = kT*rho/(e*tanh(e/(2.0*kT)))
    return trapz(integrand, e)
```

```python
dw = debye_waller_integral(e,rho_no_diff2, k*T)
wc = trapz(rho_no_diff2, e)
print "Debye Waller integral =", dw
print "Normalization of the continuous spectrum =", wc
print "Normalization of translational spectrum =", wt
print "Sum = ", wc + wt
```
